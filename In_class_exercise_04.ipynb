```python
# Write your code here
#1.1
import nltk
import pandas
import numpy
from nltk.corpus import stopwords
nltk.download('stopwords')
stopw = stopwords.words('english')

file = open('/content/InClassExercise4 (1).txt', "r", encoding="utf8")
data = file.read()
words = data.split()
print('num of words:', len(words))
sentences = data.split('.')
print('num of sentences:', sentences)
print('num of characters:', len(data))
avg = sum(len(word) for word in words)
print('avg word length:', avg)

stopwordsCount = 0;
SpecialCharCount = 0;
numericsCount = 0;
UppercaseCount = 0;

for word in words:
  if word in stopw:
    stopwordsCount += 1
  for i in range(len(word)):
    if(word[i].isalpha):
      continue
    elif(word[i].isdigit):
      numericsCount += 1
    else:
      SpecialCharCount += 1
  if word.isupper:
    UppercaseCount += 1


print('Num of stop words are:', stopwordsCount)
print('Num of special characters are:', SpecialCharCount)
print('Num of numeric digits are:', numericsCount)
print('Num of upper case words are:', UppercaseCount)
```

## 2.1 Write a Python program to remove leading zeros from an IP address. (4 points)

## 2.2 Write a Python Program to extract all the years from the following sentence. (4 points)

sentence = "The 2010s were a dramatic decade, filled with ups and downs, more than 1000 stroies have happened. As the decade comes to a close, Insider took a look back at some of the biggest headline-grabbing stories, from 2010 to 2019.
```python
# Write your code here

import re
sentence = "The 2010s were a dramatic decade, filled with ups and downs, more than 1000 stroies have happened. As the decade comes to a close, Insider took a look back at some of the biggest headline-grabbing stories, from 2010 to 2019. The result was 119 news stories that ranged from the heartwarming rescue of a Thai boys' soccer team from a flooded cave to the divisive election of President Donald Trump"
result = re.findall(r'2\d{3}', sentence)
print(result)
```
