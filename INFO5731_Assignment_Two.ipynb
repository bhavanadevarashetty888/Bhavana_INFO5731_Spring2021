# **Question 2**

(30 points). Write a python program to **clean the text data** you collected above and save the data in a new column in the csv file. The data cleaning steps include:

(1) Remove noise, such as special characters and punctuations.

(2) Remove numbers.

(3) Remove stopwords by using the [stopwords list](https://gist.github.com/sebleier/554280).

(4) Lowercase all texts

(5) Stemming. 

(6) Lemmatization.

```python
# Write your code here
import string
punctuations = dataframe['reviews'].apply(lambda j:''.join([i for i in j if i not in string.punctuation])).to_frame().rename(columns={'reviews':'punctuations'})

punct_specialchar = punctuations['punctuations'].apply(lambda j:''.join([i for i in j if i.isalnum() or i == " "])).to_frame().rename(columns={"punctuations":"Punctuations and special characters"})


dataframe1 = pandas.concat([dataframe, punct_specialchar], axis=1)
dataframe1
              



```

```python
#2.2 Removing numbers

numbers_df = dataframe1['Punctuations and special characters'].apply(lambda x:''.join([i for i in x if i == " " or i.isalpha()])).to_frame().rename(columns={"Punctuations and special characters":"numbers"})
dataframe2 = pandas.concat([dataframe1,numbers_df],axis=1)
dataframe2

```

```python
#2.3 Removing stop words
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
stpwords = set(stopwords.words('english'))
```

```python
stop_words_dataframe =  dataframe2['numbers'].apply(lambda j:' '.join([i for i in j.split() if i not in stpwords])).to_frame().rename(columns={"numbers":"stop_words"})
dataframe3 = pandas.concat([dataframe2,stop_words_dataframe],axis=1)
dataframe3
```

```python
# lower case 

lower_case_dataframe =  dataframe3['stop_words'].apply(lambda j: j.lower()).to_frame().rename(columns={"stop_words":"lowerCase"})
dataframe4 = pandas.concat([dataframe3,lower_case_dataframe],axis=1)
dataframe4
```

```python
# stemming

nltk.download('punkt')
```

```python
from nltk.stem import PorterStemmer 
from nltk.tokenize import word_tokenize 
   
ps = PorterStemmer()

tokenisation = dataframe4["lowerCase"].apply(lambda j : word_tokenize(str(j))).to_frame().rename(columns = {"lowerCase":"tokenisation"})
stemming_dataframe = tokenisation['tokenisation'].apply(lambda j: ' '.join([ps.stem(i) for i in j])).to_frame().rename(columns={"tokenisation":"stemming"})
dataframe5 = pandas.concat([dataframe4,stemming_dataframe],axis=1)
dataframe5
```

```python
#lemmatization

nltk.download('wordnet')
```

```python
import textblob
lemmetization_dataframe = tokenisation['tokenisation'].apply(lambda j: ' '.join([textblob.Word(i).lemmatize() for i in j])).to_frame().rename(columns={"tokenisation":"lemmatization"})
dataframe6 = pandas.concat([dataframe5,lemmetization_dataframe],axis=1)
dataframe6
```
